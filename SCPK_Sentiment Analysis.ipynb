```python
#proses pencarian dengan parameter kata kunci, bahasa, dan jumlah data
searchResult = api.search(q="tokyo olympics", lang="en", count=100)
```

```python
analysisResult = [] #menyimpan variable yang dihasilkan for loop

for tweet in searchResult:
    tweet_properties = {}
    tweet_properties["Date_Created"] = tweet.created_at
    tweet_properties["User"] = tweet.user.screen_name
    tweet_properties["Tweet_Content"] = tweet.text
    tweet_clear = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",tweet.text).split())
   
    analysis = TextBlob(tweet_clear)
    #print(analysis.sentiment.polarity)
    
    if analysis.sentiment.polarity > 0.0:
        tweet_properties["sentiment"] = "Positive"
    elif analysis.sentiment.polarity == 0.0:
        tweet_properties["sentiment"] = "Neutral"
    else:
        tweet_properties["sentiment"] = "Negative"
        
    #Untuk menghilangkan property retweet
    if tweet.retweet_count > 0:
        if tweet_properties not in analysisResult:
            analysisResult.append(tweet_properties)
    else:
        analysisResult.append(tweet_properties)
```

```python
#menyimpan data sesuai dengan sentimentnya
tweet_positive = [t for t in analysisResult if t["sentiment"]=="Positive"]
tweet_neutral = [t for t in analysisResult if t["sentiment"]=="Neutral"]
tweet_negative = [t for t in analysisResult if t["sentiment"]=="Negative"]
print(tweet_positive)
```

```python
#Statistik sentimentnya
print("Sentiment Result")
print("Positive: ",len(tweet_positive),"({}%)".format(100*len(tweet_positive)/len(analysisResult)))
print("Neutral: ",len(tweet_neutral),"({}%)".format(100*len(tweet_neutral)/len(analysisResult)))
print("Negative: ",len(tweet_negative),"({}%)".format(100*len(tweet_negative)/len(analysisResult)))
```
